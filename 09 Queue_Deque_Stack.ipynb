# **Queues Deque and Stacks**

## **Queue**






```python
import queue

q = queue.Queue()

numbers = [1, 2, 3, 4, 5, 6, 7]
for number in numbers:
    q.put(number)

print(q.get())
print(q.get())
# Nachdem ein Element aus der Queue geholt wurde, ist dieses Element nicht mehr in der Queue enthalten.
```

```python
# es kann mit LifoQueue ein Queue Objekt erzeugt werden, bei dem immer die letzten Elemente der Queue abgerufen werden
# zusätzlich kann eine Priorität von 0 bis 100 mit ProirityQueue übergeben werden

pq = queue.PriorityQueue()

pq.put((2, "Hello World"))
pq.put((7, 99))
pq.put((1, "First Element"))
pq.put((43, 4.3))
while not pq.empty():
    print(pq.get())
```

## **Deque Double Ended Queue**





```python
from collections import deque

people = ["Mario", "Luigi", "Kirby", "Sonic", "Tails"]

dq = deque(people)
dq.append("Bowser")
print(dq)

dq.popleft()
print(dq)
dq.appendleft("Robotnik")
print(dq)

dq.rotate(-1)
print(dq)

# Imagine People are coming in loads
dq.extend(["Shy Guy", "Knuckles"])
print(dq)
dq.extendleft(["Link","Zelda"])
print(dq)

dq.reverse()
print(dq)
```
