# **Array Lists vs Linked Lists**

## **Array List**
unflexibel\
zu benutzen bei häufigem Auslesen



## **Linked List**
flexibler als Arrays\
zu benutzen wenn die Reihenfolge der Elemente, innerhalb der Liste, häufig verändert wird



ersteinmal nur das Startobjekt sichtbar\
beim Array sind alle Elemente sofort sichtbar

Bei Änderungen in einer LinkedList werden die Referenzen bzw. Pointer einfach auf neue Speicherorte umgebogen.\
Langsame Suche deshalb, weil die LinkedList immer von Beginn jedes seiner Elemente anfassen muss,\
damit die Information für das nächste Element ausgelesen werden kann.



## **Double Linked List**
hat noch eine Verknüpfung zum Vorgänger und besitzt zusätzlich noch ein End Element\
dadurch kann von beiden Seiten mit der Suche begonnen werden -> bessere Performance\
mehr Verwaltung da die reverse Pointer erfasst werden müssen



```python
# How to create a Linked List in Python

# Subclass of LinkedList
class node:
    def __init__(self, data= None):
        self.data = data
        self.next = None

class linked_list:
    def __init__(self):
        self.head = node()
    
    def append(self, data):
        new_node = node(data)
        current_node = self.head
        while current_node.next != None:
            current_node = current_node.next
        current_node.next = new_node
    
    def length(self):
        current_node = self.head
        total_nodes = 0
        while current_node.next!= None:
            total_nodes += 1
            current_node = current_node.next
        return total_nodes
    
    def display(self):
        elements = []
        current_nodes = self.head
        while current_nodes.next!= None:
            current_nodes = current_nodes.next
            elements.append(current_nodes.data)
        print(elements)
        
my_list = linked_list()
my_list.append(100)
my_list.append("String")
my_list.display()
```
